```python
import rasterio
from matplotlib import pyplot
import os
```

```python
path_mos_ppt = 's3://dev-et-data/in/USA_data/precip_gridmet_tiffs/2000/prec_2000003.tif' 
 
src = rasterio.open(path_mos_ppt)
pyplot.imshow(src.read(1), cmap='pink')
pyplot.show()
```

```python
from rasterio.plot import show_hist
show_hist(src, bins=50, lw=0.0, stacked=False, alpha=0.3, histtype='stepfilled', title="Histogram")
```
