```python
def _mkdir(directory):
    try:
        os.stat(directory)
    except:
        os.mkdir(directory) 
```

```python
_mkdir('./nc_wzell')
```

```python
def create_s3_list_of_months(main_prefix, year, output_name):
    the_list = []
    for i in range(1,13):
        month = f'{i:02d}'
        file_object = main_prefix + output_name + str(year) + month + '.tif'
        the_list.append(file_object)
    return the_list
```

```python
years = range(2003,2016)
main_prefix = './junkbox/'
output_name='ppt_avg_'
for year in years:
    my_tifs = create_s3_list_of_months(main_prefix, year, output_name)
    ds = xr_build_cube_concat_ds(my_tifs, output_name)
    nc_file_name = './nc_wzell' +'/' + output_name + str(year) + '.nc'
    ds.to_netcdf(nc_file_name, engine='h5netcdf')
```
